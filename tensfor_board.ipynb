```python
import os
import datetime
# Load the TensorBoard notebook extension
%load_ext tensorboard
import tensorflow as tf
```

```python
logdir = os.path.join("logs", datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
tensorboard_callback = tf.keras.callbacks.TensorBoard(logdir, histogram_freq=1)
```

```python
import pandas as pd
df = pd.read_csv('Eda_OK.csv')
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['loan_status','installment'],axis=1).values 
y = df['loan_status'].values
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42
                                                    ,stratify = y)
from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam


model = Sequential()
model.add(Dense(70,activation='relu')) #Input Layer and HIdden Layer 1
model.add(Dense(35,activation='relu')) #HIdden Layer 3
model.add(Dense(18,activation='relu')) #HIdden Layer 3
model.add(Dense(1,activation='sigmoid')) #Outpur Layer

opt=Adam(lr=0.003)
model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x=X_train,
          y=y_train,
          epochs=25,
          batch_size=256,
          validation_data=(X_test, y_test),
          callbacks=[tensorboard_callback]
          )
```
